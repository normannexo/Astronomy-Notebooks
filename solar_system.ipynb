### Solar system

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

from datetime import datetime
from astropy.time import Time 
from astropy.coordinates import solar_system_ephemeris, EarthLocation 
from astropy.coordinates import get_body_barycentric, get_body, get_moon
from astroquery.jplhorizons import Horizons
```

```python
t = Time.now()
loc = EarthLocation.of_site('greenwich')
with solar_system_ephemeris.set('builtin'):
    jupiter = get_body('jupiter', t, loc)
cart_earth = get_body_barycentric('sun', t)
```

```python
objects = ['sun', 'earth', 'jupiter', 'mars', 'venus', 'mercury', 'saturn']

objs = []
for obj in objects:
    coords = get_body_barycentric(obj, t)
    objs.append({'name':obj, "x":coords.x.value, "y":coords.y.value })
objs
df_objs = pd.DataFrame(objs)
df_objs
```

```python
sns.scatterplot(data = df_objs, x = 'x', y='y', hue = 'name')
```
